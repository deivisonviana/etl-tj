```python
## TB_DIM_CARGO
from pyspark.sql import DataFrame
from delta.tables import DeltaTable
from pyspark.sql.utils import AnalysisException
from pyspark.sql import Window
from pyspark.sql import functions as F
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, LongType
from pyspark.sql.functions import col, lit, monotonically_increasing_id, row_number
import shutil
import os

def deduplicar_e_adicionar_ids_e_salvar(df_novo, delta_path, id_col="id_cargo", dedup_cols=["ds_cargo"]):
    """
    Deduplica registros do DataFrame novo comparando com dados existentes no Delta Lake,
    adiciona coluna de id incremental e salva no Delta em modo append.

    :param df_novo: DataFrame com os novos dados
    :param delta_path: Caminho da tabela Delta (ex: tb_dim_tarefas_slv)
    :param id_col: Nome da coluna de ID incremental (default: "id_tarefa")
    :param dedup_cols: Lista de colunas usadas para deduplicação (default: ["ds_tarefa"])
    :return: DataFrame resultante (com coluna de id gerada ou preservada)
    """
    print(f"Iniciando processo para salvar dados no Delta: {delta_path}")
    
    df_existente = None
    max_id = 0
    
    # Tenta ler dados existentes para deduplicação e encontrar o maior ID
    try:
        if DeltaTable.isDeltaTable(spark, delta_path):
            df_existente = spark.read.format("delta").load(delta_path)
            print("Delta carregado para deduplicação.")
            
            # Buscar maior id já existente
            # Verifica se a coluna id_col existe antes de tentar buscar o máximo
            if id_col in df_existente.columns:
                max_id_row = df_existente.agg(F.max(id_col).alias("max_id")).collect()[0]
                max_id = max_id_row["max_id"] if max_id_row["max_id"] is not None else 0
            else:
                print(f"Coluna '{id_col}' não encontrada no Delta existente. Iniciando ID do zero.")
                max_id = 0
            
            print(f"max_id atual: {max_id}")
        else:
            print(f"Tabela Delta em {delta_path} não existe. Todos os registros serão considerados novos.")
            max_id = 0
    except Exception as e:
        print(f"Erro ao ler ou verificar Delta existente em {delta_path}: {e}")
        print("Todos os registros serão considerados novos e o ID começará do zero.")
        max_id = 0

    # Deduplicação: pega apenas registros novos
    df_novos_para_processar = df_novo
    if df_existente is not None:
        df_existente_chave = df_existente.select(*dedup_cols).dropDuplicates()
        df_novos_para_processar = df_novo.join(
            df_existente_chave,
            on=dedup_cols,
            how="left_anti"
        )
    
    print(f"Número de registros novos após deduplicação: {df_novos_para_processar.count()}")
    
    # Se não há novos registros para gerar ID, retorna o DataFrame original com id existente
    # ou um DataFrame vazio se a entrada for vazia.
    if df_novos_para_processar.count() == 0:
        print("Nenhum registro novo para salvar ou gerar ID. Retornando DataFrame existente ou vazio.")
        # Se existe df_existente, retorna ele, caso contrário, retorna um df_novo vazio com schema correto
        if df_existente is not None:
             # Garante que a coluna id_col esteja presente no df_existente
            if id_col not in df_existente.columns:
                df_existente = df_existente.withColumn(id_col, lit(None).cast(LongType()))
            return df_existente
        else:
            # Cria um DataFrame vazio com as colunas esperadas, incluindo id_col
            return spark.createDataFrame([], schema=df_novo.schema.add(id_col, LongType()))


    # Gerar novos ids contínuos para os registros *realmente novos*
    windowSpec = Window.orderBy(F.lit(1)) # Ordem arbitrária para gerar IDs sequenciais
    df_com_id_gerado = df_novos_para_processar.withColumn(
        id_col, 
        F.row_number().over(windowSpec).cast(LongType()) + F.lit(max_id).cast(LongType())
    )
    
    print(f"Deduplicação e geração de IDs finalizada. Novos IDs a partir de: {max_id + 1}")
    
    # Se houver dados existentes, unimos os novos com os existentes
    if df_existente is not None:
        # Garante que df_existente tem a coluna id_col antes do union
        if id_col not in df_existente.columns:
            df_existente = df_existente.withColumn(id_col, lit(None).cast(LongType()))
        
        # Seleciona as mesmas colunas na mesma ordem para garantir o UNION
        # Assume que df_com_id_gerado já tem todas as colunas necessárias + id_col
        cols_to_union = [id_col] + [c for c in df_com_id_gerado.columns if c != id_col]
        
        # Filtra registros existentes que não foram "antijoin" para incluir no resultado
        df_existente_nao_duplicado = df_existente.join(
            df_novos_para_processar.select(*dedup_cols), # Usamos as colunas de dedup dos novos
            on=dedup_cols,
            how="left_anti"
        ).select(cols_to_union) # Garante que as colunas são as mesmas
        
        df_resultado = df_existente_nao_duplicado.unionByName(df_com_id_gerado.select(cols_to_union))
    else:
        df_resultado = df_com_id_gerado
        
    return df_resultado




nome_workspace = notebookutils.runtime.context.get("currentWorkspaceName")

#Mapeamento dos caminhos
tb_dim_cargo_bnz = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Tables/Base/tb_dim_cargo'
tb_stg_cargo = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Base/siarhes_cargo.csv'
tb_dim_cargo_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Base/tb_dim_cargo'
pasta_origem = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Base'
pasta_lidos = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Base/Lidos'


#CARREGAR PARA CAMADA BRONZE

df_cargo = spark.read.option("header", "true").csv(tb_stg_cargo)


# Verificar se o DataFrame não está vazio
if df_cargo.rdd.isEmpty():
    print("Nenhum arquivo CSV encontrado na pasta ou CSV está vazio.")
    schema_esperado = StructType([
        StructField("codigo", StringType(), True),
        StructField("nome", StringType(), True),
        StructField("dtatualiza", StringType(), True)
    ])
    df_cargo = spark.createDataFrame([], schema=schema_esperado)
else:
    print("Arquivos CSV carregados com sucesso!")
    # Seleção e casting das colunas
    df_cargo = df_cargo.select(
        col("codigo").cast(StringType()),
        col("nome").cast(StringType()),
        col("dtatualiza").cast(StringType())
    )



# Selecionar colunas de interesse
colunas_dimensao = ['nome', 'codigo']
df_dim_cargo = df_cargo.select(*colunas_dimensao)

# Renomear colunas
df_dim_cargo = df_dim_cargo.select(
    col("codigo").alias("cd_sistema_origem"),
    col("nome").alias("ds_cargo")
)


# Verifica se o DataFrame de dimensão não está vazio antes de tentar gravar na Bronze
if not df_dim_cargo.rdd.isEmpty():
    # Verifica se a DeltaTable já existe na camada Bronze
    if DeltaTable.isDeltaTable(spark, tb_dim_cargo_bnz):
        delta_table = DeltaTable.forPath(spark, tb_dim_cargo_bnz)

        # Realiza o MERGE (upsert)
        delta_table.alias("target") \
            .merge(
                source=df_dim_cargo.alias("source"),
                condition="target.cd_sistema_origem = source.cd_sistema_origem"
            ) \
            .whenMatchedUpdateAll().whenNotMatchedInsertAll().execute()
    else:
        # Se a tabela não existir, cria no formato Delta
        df_dim_cargo.write.format("delta").mode("append").option("overwriteSchema", "true").save(tb_dim_cargo_bnz)
else:
    print("DataFrame de dimensão vazio. Nenhuma gravação na camada Bronze.")

   
# CARREGAR PARA CAMADA SILVER

df_dim_cargo = spark.read.format("delta").load(tb_dim_cargo_bnz)


# Adiciona a linha "NÃO DETERMINADO"
schema = df_dim_cargo.schema
df_novo = spark.createDataFrame(
    [(0, "NAO DETERMINADO")],
    ["cd_sistema_origem", "ds_cargo"]
)
df_dim_cargo = df_novo.union(df_dim_cargo)

# # Cria a coluna id_cargo com valores sequenciais (monotonic workaround)
# window_spec = Window.orderBy(lit(1))
# df_dim_cargo = df_dim_cargo.withColumn("id_cargo", row_number().over(window_spec).cast("long"))

# Elimina registro repetido e cria coluna id_tarefa começando do max_id + 1
df_dim_cargo = deduplicar_e_adicionar_ids_e_salvar(
    df_novo=df_dim_cargo,
    delta_path=tb_dim_cargo_slv,
    id_col="id_cargo",
    dedup_cols=["cd_sistema_origem"]   # lista de colunas que definem duplicidade
)



# Reordena colunas para deixar id_cargo no início
colunas_ordenadas = ["id_cargo"] + [col for col in df_dim_cargo.columns if col != "id_cargo"]
df_dim_cargo = df_dim_cargo.select(*colunas_ordenadas)


# Verifica se o DataFrame de dimensão não está vazio antes de tentar gravar na Prata
if not df_dim_cargo.rdd.isEmpty():
    # Verifica se a tabela já existe e faz merge ou write
    if DeltaTable.isDeltaTable(spark, tb_dim_cargo_slv):
        TableDimCargo = DeltaTable.forPath(spark, tb_dim_cargo_slv)

        TableDimCargo.alias("target").merge(
            source=df_dim_cargo.alias("source"),
            condition="target.cd_sistema_origem = source.cd_sistema_origem"
        ).whenMatchedUpdateAll().whenNotMatchedInsertAll().execute()

        # print("merge deltatablelake 2 TB_DIM_CARGO")
    else:
        df_dim_cargo.write.format("delta").mode("append").save(tb_dim_cargo_slv)
        # print("Write deltatablelake 2 TB_DIM_CARGO")
else:
    print("DataFrame de dimensão vazio. Nenhuma gravação na camada Prata.")





#CARREGAR PARA CAMADA OURO

tb_dim_cargo_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Base/tb_dim_cargo'
# tb_dim_cargo_slv = 'LH_SLV_Produtividade.Base.tb_dim_cargo'
tb_dim_cargo_gld_path = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_GLD_Produtividade.Lakehouse/Tables/Base/tb_dim_cargo'
tb_dim_cargo_gld = 'LH_GLD_Produtividade.Base.tb_dim_cargo' 

#tb_dim_cargo
if spark.catalog._jcatalog.tableExists(tb_dim_cargo_gld):
    spark.sql(f"DROP TABLE {tb_dim_cargo_gld}")
    print(f"✅ Tabela '{tb_dim_cargo_gld}' foi excluída.")
else:
    print(f"ℹ️ Tabela '{tb_dim_cargo_gld}' não existe.")

dt = DeltaTable.forPath(spark, tb_dim_cargo_slv)
df_dados = dt.toDF()
df_dados.write.format("delta").mode("overwrite").save(tb_dim_cargo_gld_path)



```

```python
##10 DIM_TB_VINCULO

from pyspark.sql.window import Window
from delta.tables import DeltaTable
from pyspark.sql.utils import AnalysisException
from pyspark.sql import functions as F
from pyspark.sql.functions import col, lit, monotonically_increasing_id, row_number
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, LongType
import shutil
import os

def deduplicar_e_adicionar_ids_e_salvar(df_novo, delta_path, id_col="id_vinculo", dedup_cols=["ds_vinculo"]):
    """
    Deduplica registros do DataFrame novo comparando com dados existentes no Delta Lake,
    adiciona coluna de id incremental e salva no Delta em modo append.

    :param df_novo: DataFrame com os novos dados
    :param delta_path: Caminho da tabela Delta (ex: tb_dim_tarefas_slv)
    :param id_col: Nome da coluna de ID incremental (default: "id_tarefa")
    :param dedup_cols: Lista de colunas usadas para deduplicação (default: ["ds_tarefa"])
    :return: DataFrame resultante (com coluna de id gerada ou preservada)
    """
    print(f"Iniciando processo para salvar dados no Delta: {delta_path}")
    
    df_existente = None
    max_id = 0
    
    # Tenta ler dados existentes para deduplicação e encontrar o maior ID
    try:
        if DeltaTable.isDeltaTable(spark, delta_path):
            df_existente = spark.read.format("delta").load(delta_path)
            print("Delta carregado para deduplicação.")
            
            # Buscar maior id já existente
            # Verifica se a coluna id_col existe antes de tentar buscar o máximo
            if id_col in df_existente.columns:
                max_id_row = df_existente.agg(F.max(id_col).alias("max_id")).collect()[0]
                max_id = max_id_row["max_id"] if max_id_row["max_id"] is not None else 0
            else:
                print(f"Coluna '{id_col}' não encontrada no Delta existente. Iniciando ID do zero.")
                max_id = 0
            
            print(f"max_id atual: {max_id}")
        else:
            print(f"Tabela Delta em {delta_path} não existe. Todos os registros serão considerados novos.")
            max_id = 0
    except Exception as e:
        print(f"Erro ao ler ou verificar Delta existente em {delta_path}: {e}")
        print("Todos os registros serão considerados novos e o ID começará do zero.")
        max_id = 0

    # Deduplicação: pega apenas registros novos
    df_novos_para_processar = df_novo
    if df_existente is not None:
        df_existente_chave = df_existente.select(*dedup_cols).dropDuplicates()
        df_novos_para_processar = df_novo.join(
            df_existente_chave,
            on=dedup_cols,
            how="left_anti"
        )
    
    print(f"Número de registros novos após deduplicação: {df_novos_para_processar.count()}")
    
    # Se não há novos registros para gerar ID, retorna o DataFrame original com id existente
    # ou um DataFrame vazio se a entrada for vazia.
    if df_novos_para_processar.count() == 0:
        print("Nenhum registro novo para salvar ou gerar ID. Retornando DataFrame existente ou vazio.")
        # Se existe df_existente, retorna ele, caso contrário, retorna um df_novo vazio com schema correto
        if df_existente is not None:
             # Garante que a coluna id_col esteja presente no df_existente
            if id_col not in df_existente.columns:
                df_existente = df_existente.withColumn(id_col, lit(None).cast(LongType()))
            return df_existente
        else:
            # Cria um DataFrame vazio com as colunas esperadas, incluindo id_col
            return spark.createDataFrame([], schema=df_novo.schema.add(id_col, LongType()))


    # Gerar novos ids contínuos para os registros *realmente novos*
    windowSpec = Window.orderBy(F.lit(1)) # Ordem arbitrária para gerar IDs sequenciais
    df_com_id_gerado = df_novos_para_processar.withColumn(
        id_col, 
        F.row_number().over(windowSpec).cast(LongType()) + F.lit(max_id).cast(LongType())
    )
    
    print(f"Deduplicação e geração de IDs finalizada. Novos IDs a partir de: {max_id + 1}")
    
    # Se houver dados existentes, unimos os novos com os existentes
    if df_existente is not None:
        # Garante que df_existente tem a coluna id_col antes do union
        if id_col not in df_existente.columns:
            df_existente = df_existente.withColumn(id_col, lit(None).cast(LongType()))
        
        # Seleciona as mesmas colunas na mesma ordem para garantir o UNION
        # Assume que df_com_id_gerado já tem todas as colunas necessárias + id_col
        cols_to_union = [id_col] + [c for c in df_com_id_gerado.columns if c != id_col]
        
        # Filtra registros existentes que não foram "antijoin" para incluir no resultado
        df_existente_nao_duplicado = df_existente.join(
            df_novos_para_processar.select(*dedup_cols), # Usamos as colunas de dedup dos novos
            on=dedup_cols,
            how="left_anti"
        ).select(cols_to_union) # Garante que as colunas são as mesmas
        
        df_resultado = df_existente_nao_duplicado.unionByName(df_com_id_gerado.select(cols_to_union))
    else:
        df_resultado = df_com_id_gerado
        
    return df_resultado




nome_workspace = notebookutils.runtime.context.get("currentWorkspaceName")

tb_dim_vinculo_bnz = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Tables/Base/tb_dim_vinculo'
tb_stg_vinculo = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Base/siarhes_vinculo.csv'
tb_dim_vinculo_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Base/tb_dim_vinculo'
pasta_origem = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Base'
pasta_lidos = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Base/Lidos'


## Criar e dar carga da tb_dim_vinculo em bronze
df_vinculo = spark.read.option("header", "true").csv(tb_stg_vinculo)

# Verificar se o DataFrame não está vazio
if df_vinculo.rdd.isEmpty():
    print("Nenhum arquivo CSV encontrado na pasta ou CSV está vazio.")
    schema_esperado = StructType([
        StructField("tipovinc", StringType(), True),
        StructField("categoria", StringType(), True)
    ])
    df_vinculo = spark.createDataFrame([], schema=schema_esperado)
else:
    print("Arquivos CSV carregados com sucesso!")
    # Seleção e casting das colunas
    df_vinculo = df_vinculo.select(
        col("tipovinc").cast(StringType()),
        col("categoria").cast(StringType())
    )



colunas_dimensao = ['tipovinc']
df_dim_vinculo = df_vinculo.select(colunas_dimensao).dropDuplicates()
df_dim_vinculo = df_dim_vinculo.withColumnRenamed("tipovinc", "ds_vinculo")


# Verifica se o DataFrame de dimensão não está vazio antes de tentar gravar na Bronze
if not df_dim_vinculo.rdd.isEmpty():
    # Verifica se a DeltaTable já existe
    if DeltaTable.isDeltaTable(spark, tb_dim_vinculo_bnz):
        delta_table = DeltaTable.forPath(spark, tb_dim_vinculo_bnz)

        # Realiza o MERGE (upsert)
        delta_table.alias("target") \
            .merge(
                source=df_dim_vinculo.alias("source"),
                condition="target.ds_vinculo = source.ds_vinculo"
            ) \
            .whenMatchedUpdateAll().whenNotMatchedInsertAll().execute()
    else:
        # Se a tabela não existir, cria no formato Delta
        df_dim_vinculo.write.format("delta").mode("append").option("overwriteSchema", "true").save(tb_dim_vinculo_bnz)
else:
    print("DataFrame de dimensão vazio. Nenhuma gravação na camada Bronze.")

    



# CARREGAR PARA CAMADA SILVER

df_dim_vinculo = spark.read.format("delta").load(tb_dim_vinculo_bnz)
table_path = tb_dim_vinculo_slv

# Criação da nova linha
df_nova_linha = spark.createDataFrame([{'ds_vinculo': 'NAO DETERMINADO'}])

# Concatenar nova linha + df original
df_dim_vinculo = df_nova_linha.unionByName(df_dim_vinculo)

# Criar coluna id_vinculo sequencial
# windowSpec = Window.orderBy(lit(1))
# df_dim_vinculo = df_dim_vinculo.withColumn("id_vinculo", row_number().over(windowSpec).cast("double"))

# Elimina registro repetido e cria coluna id_tarefa começando do max_id + 1
df_dim_vinculo = deduplicar_e_adicionar_ids_e_salvar(
    df_novo=df_dim_vinculo,
    delta_path=tb_dim_vinculo_slv,
    id_col="id_vinculo",
    dedup_cols=["ds_vinculo"]   # lista de colunas que definem duplicidade
)




# Reordena colunas para deixar id_vinculo no início
colunas_ordenadas = ["id_vinculo"] + [col for col in df_dim_vinculo.columns if col != "id_vinculo"]
df_dim_vinculo = df_dim_vinculo.select(*colunas_ordenadas)

# Verifica se o DataFrame de dimensão não está vazio antes de tentar gravar na Bronze
if not df_dim_vinculo.rdd.isEmpty():
    # Fazer merge com Delta Lake
    if DeltaTable.isDeltaTable(spark, tb_dim_vinculo_slv):
        deltaTable = DeltaTable.forPath(spark, tb_dim_vinculo_slv)
        deltaTable.alias("target").merge(
            source=df_dim_vinculo.alias("source"),
            condition="target.ds_vinculo = source.ds_vinculo"
        ).whenMatchedUpdateAll().whenNotMatchedInsertAll().execute()
        print("merge deltatablelake 4 TB_DIM_VINCULO")
    else:
        df_dim_vinculo.write.format("delta").mode("append").save(tb_dim_vinculo_slv)
        print("Write deltatablelake 4 TB_DIM_VINCULO")
else:
    print("DataFrame de dimensão vazio. Nenhuma gravação na camada Bronze.")    

# df_dim_vinculo.show(5)






#CARREGAR PARA CAMADA OURO

tb_dim_vinculo_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Base/tb_dim_vinculo'
# tb_dim_vinculo_slv = 'LH_SLV_Produtividade.Base.tb_dim_vinculo'
tb_dim_vinculo_gld_path = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_GLD_Produtividade.Lakehouse/Tables/Base/tb_dim_vinculo'
tb_dim_vinculo_gld = 'LH_GLD_Produtividade.Base.tb_dim_vinculo' 


#tb_dim_vinculo
if spark.catalog._jcatalog.tableExists(tb_dim_vinculo_gld):
    spark.sql(f"DROP TABLE {tb_dim_vinculo_gld}")
    print(f"✅ Tabela '{tb_dim_vinculo_gld}' foi excluída.")
else:
    print(f"ℹ️ Tabela '{tb_dim_vinculo_gld}' não existe.")


dt = DeltaTable.forPath(spark, tb_dim_vinculo_slv)
df_dados = dt.toDF()
table_path = tb_dim_vinculo_gld_path
df_dados.write.format("delta").mode("overwrite").save(table_path)

```

```python
##12DIM_TB_COMARCA

from pyspark.sql.window import Window
from delta.tables import DeltaTable
from pyspark.sql.utils import AnalysisException
from pyspark.sql.functions import col, when, rank, lit
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, LongType
from pyspark.sql.types import BooleanType
from pyspark.sql.types import LongType
from pyspark.sql import functions as F
import shutil
import os

def deduplicar_e_adicionar_ids_e_salvar(df_novo, delta_path, id_col="id_comarca", dedup_cols=["ds_comarca"]):
    """
    Deduplica registros do DataFrame novo comparando com dados existentes no Delta Lake,
    adiciona coluna de id incremental e salva no Delta em modo append.

    :param df_novo: DataFrame com os novos dados
    :param delta_path: Caminho da tabela Delta (ex: tb_dim_tarefas_slv)
    :param id_col: Nome da coluna de ID incremental (default: "id_tarefa")
    :param dedup_cols: Lista de colunas usadas para deduplicação (default: ["ds_tarefa"])
    :return: DataFrame resultante (com coluna de id gerada ou preservada)
    """
    print(f"Iniciando processo para salvar dados no Delta: {delta_path}")
    
    df_existente = None
    max_id = 0
    
    # Tenta ler dados existentes para deduplicação e encontrar o maior ID
    try:
        if DeltaTable.isDeltaTable(spark, delta_path):
            df_existente = spark.read.format("delta").load(delta_path)
            print("Delta carregado para deduplicação.")
            
            # Buscar maior id já existente
            # Verifica se a coluna id_col existe antes de tentar buscar o máximo
            if id_col in df_existente.columns:
                max_id_row = df_existente.agg(F.max(id_col).alias("max_id")).collect()[0]
                max_id = max_id_row["max_id"] if max_id_row["max_id"] is not None else 0
            else:
                print(f"Coluna '{id_col}' não encontrada no Delta existente. Iniciando ID do zero.")
                max_id = 0
            
            print(f"max_id atual: {max_id}")
        else:
            print(f"Tabela Delta em {delta_path} não existe. Todos os registros serão considerados novos.")
            max_id = 0
    except Exception as e:
        print(f"Erro ao ler ou verificar Delta existente em {delta_path}: {e}")
        print("Todos os registros serão considerados novos e o ID começará do zero.")
        max_id = 0

    # Deduplicação: pega apenas registros novos
    df_novos_para_processar = df_novo
    if df_existente is not None:
        df_existente_chave = df_existente.select(*dedup_cols).dropDuplicates()
        df_novos_para_processar = df_novo.join(
            df_existente_chave,
            on=dedup_cols,
            how="left_anti"
        )
    
    print(f"Número de registros novos após deduplicação: {df_novos_para_processar.count()}")
    
    # Se não há novos registros para gerar ID, retorna o DataFrame original com id existente
    # ou um DataFrame vazio se a entrada for vazia.
    if df_novos_para_processar.count() == 0:
        print("Nenhum registro novo para salvar ou gerar ID. Retornando DataFrame existente ou vazio.")
        # Se existe df_existente, retorna ele, caso contrário, retorna um df_novo vazio com schema correto
        if df_existente is not None:
             # Garante que a coluna id_col esteja presente no df_existente
            if id_col not in df_existente.columns:
                df_existente = df_existente.withColumn(id_col, lit(None).cast(LongType()))
            return df_existente
        else:
            # Cria um DataFrame vazio com as colunas esperadas, incluindo id_col
            return spark.createDataFrame([], schema=df_novo.schema.add(id_col, LongType()))


    # Gerar novos ids contínuos para os registros *realmente novos*
    windowSpec = Window.orderBy(F.lit(1)) # Ordem arbitrária para gerar IDs sequenciais
    df_com_id_gerado = df_novos_para_processar.withColumn(
        id_col, 
        F.row_number().over(windowSpec).cast(LongType()) + F.lit(max_id).cast(LongType())
    )
    
    print(f"Deduplicação e geração de IDs finalizada. Novos IDs a partir de: {max_id + 1}")
    
    # Se houver dados existentes, unimos os novos com os existentes
    if df_existente is not None:
        # Garante que df_existente tem a coluna id_col antes do union
        if id_col not in df_existente.columns:
            df_existente = df_existente.withColumn(id_col, lit(None).cast(LongType()))
        
        # Seleciona as mesmas colunas na mesma ordem para garantir o UNION
        # Assume que df_com_id_gerado já tem todas as colunas necessárias + id_col
        cols_to_union = [id_col] + [c for c in df_com_id_gerado.columns if c != id_col]
        
        # Filtra registros existentes que não foram "antijoin" para incluir no resultado
        df_existente_nao_duplicado = df_existente.join(
            df_novos_para_processar.select(*dedup_cols), # Usamos as colunas de dedup dos novos
            on=dedup_cols,
            how="left_anti"
        ).select(cols_to_union) # Garante que as colunas são as mesmas
        
        df_resultado = df_existente_nao_duplicado.unionByName(df_com_id_gerado.select(cols_to_union))
    else:
        df_resultado = df_com_id_gerado
        
    return df_resultado





nome_workspace = notebookutils.runtime.context.get("currentWorkspaceName")

tb_dim_comarca_bnz = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Tables/Base/tb_dim_comarca'
tb_stg_setor = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Base/siarhes_setor.csv' #Setor -> Comarca, unidade e tipo_unidade
pasta_origem = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Base'
pasta_lidos = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Base/Lidos'

## Criar e dar carga da tb_dim_comarca em bronze
df_comarca = spark.read.option("header", "true").csv(tb_stg_setor)

# Verificar se o DataFrame não está vazio
if df_comarca.rdd.isEmpty():
    print("Nenhum arquivo CSV encontrado na pasta ou CSV está vazio.")
    schema_esperado = StructType([
        StructField("setor", StringType(), True),
        StructField("nomesetor", StringType(), True),
        StructField("tiposetor", StringType(), True),
        StructField("dtini", StringType(), True),
        StructField("dtfim", StringType(), True),
        StructField("cd_orgao", StringType(), True)
    ])
    df_comarca = spark.createDataFrame([], schema=schema_esperado)
else:
    print("Arquivos CSV carregados com sucesso!")
    # Seleção e casting das colunas
    df_comarca = df_comarca.select(
        col("setor").cast(StringType()),
        col("nomesetor").cast(StringType()),
        col("tiposetor").cast(StringType()),         
        col("dtini").cast(StringType()),
        col("dtfim").cast(StringType()),
        col("cd_orgao").cast(StringType())
    )
    

# 1. Filtro para manter apenas registros do tipo COMARCA
df_comarca_historico = df_comarca.filter(col("tiposetor") == "COMARCA")

# 2. Criação da coluna 'ativo'
df_comarca_historico = df_comarca_historico.withColumn("ativo", col("dtfim").isNull())

# 3. Seleção de colunas
colunas_comarca = ["setor", "nomesetor", "tiposetor", "ativo", "dtini", "cd_orgao"]
df_comarca_historico = df_comarca_historico.select(*colunas_comarca)

# 4. Rank reverso por 'setor' e 'dtini'
window_spec = Window.partitionBy("setor").orderBy(col("dtini").desc())
df_comarca_historico = df_comarca_historico.withColumn(
    "versao_ordem_inversa", rank().over(window_spec)
)

# 5. Filtro para manter somente a última versão por setor
df_comarca = df_comarca_historico.filter(col("versao_ordem_inversa") == 1)

# 6. Renomeia colunas e reordena
df_dim_comarca = df_comarca.select(
    col("setor").alias("cd_sistema_origem"),
    col("nomesetor").alias("ds_comarca"),
    col("ativo").alias("in_ativo").cast("boolean"),
    col("cd_orgao").alias("cd_corporativo")
).orderBy("ds_comarca")


# Verifica se o DataFrame de dimensão não está vazio antes de tentar gravar na Bronze
if not df_dim_comarca.rdd.isEmpty():
    # Verifica se a DeltaTable já existe
    if DeltaTable.isDeltaTable(spark, tb_dim_comarca_bnz):
        delta_table = DeltaTable.forPath(spark, tb_dim_comarca_bnz)

        # Realiza o MERGE (upsert)
        delta_table.alias("target") \
            .merge(
                source=df_dim_comarca.alias("source"),
                condition="target.cd_sistema_origem = source.cd_sistema_origem"
            ) \
            .whenMatchedUpdateAll().whenNotMatchedInsertAll().execute()
        print("merge deltatablelake 6 TB_DIM_TIPO_UNIDADE")
    else:
        # Se a tabela não existir, cria no formato Delta
        df_dim_comarca.write.format("delta").mode("overwrite").option("overwriteSchema", "true").save(tb_dim_comarca_bnz)
        print("Write deltatablelake 6 TB_DIM_TIPO_UNIDADE")
else:
    print("DataFrame de dimensão vazio. Nenhuma gravação na camada Bronze.")

    


# CARREGAR PARA CAMADA SILVER

# Caminhos
tb_dim_comarca_bnz = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Tables/Base/tb_dim_comarca'
tb_dim_comarca_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Base/tb_dim_comarca'

# Ler tabela bronze
delta_bronze = DeltaTable.forPath(spark, tb_dim_comarca_bnz)
df_dim_comarca = delta_bronze.toDF()

# # Define o schema explicitamente
# schema = StructType([
#     StructField("ds_comarca", StringType(), True),
#     StructField("in_ativo", BooleanType(), True),
#     StructField("cd_sistema_origem", LongType(), True),
#     StructField("cd_corporativo", IntegerType(), True)
# ])

# Converta df_dim_comarca para o mesmo schema
# df_dim_comarca = df_dim_comarca.select(
#     col("ds_comarca").cast("string"),
#     col("in_ativo").cast("boolean"),
#     col("cd_sistema_origem").cast("long"),
#     col("cd_corporativo").cast("int")
# )


# Cria a nova linha com schema definido
nova_linha = spark.createDataFrame(
    [(None, "NAO DETERMINADO", True, None)],
    schema=df_dim_comarca.schema
)
      


# Inserir nova linha no topo
df_dim_comarca = nova_linha.union(df_dim_comarca)

# # Gerar coluna id_comarca sequencial
# window_spec = Window.orderBy(F.lit("A"))  # Ordem arbitrária
# df_dim_comarca = df_dim_comarca.withColumn("id_comarca", F.row_number().over(window_spec))

# Elimina registro repetido e cria coluna id_tarefa começando do max_id + 1
df_dim_comarca = deduplicar_e_adicionar_ids_e_salvar(
    df_novo=df_dim_comarca,
    delta_path=tb_dim_comarca_slv,
    id_col="id_comarca",
    dedup_cols=["cd_sistema_origem"]   # lista de colunas que definem duplicidade
)




# Reorganizar colunas: id_comarca primeiro
cols = ["id_comarca"] + [c for c in df_dim_comarca.columns if c != "id_comarca"]
df_dim_comarca = df_dim_comarca.select(cols)

# Verifica se o DataFrame de dimensão não está vazio antes de tentar gravar na Prata
if not df_dim_comarca.rdd.isEmpty():    
    # Verifica se a DeltaTable já existe
    if DeltaTable.isDeltaTable(spark, tb_dim_comarca_slv):
        delta_table = DeltaTable.forPath(spark, tb_dim_comarca_slv)

        # Realiza o MERGE (upsert)
        delta_table.alias("target") \
            .merge(
                source=df_dim_comarca.alias("source"),
                condition="target.cd_sistema_origem = source.cd_sistema_origem"
            ) \
            .whenMatchedUpdateAll().whenNotMatchedInsertAll().execute()
    else:
        # Se a tabela não existir, cria no formato Delta
        df_dim_comarca.write.format("delta").mode("overwrite").option("overwriteSchema", "true").save(tb_dim_comarca_slv)
else:
    print("DataFrame de dimensão vazio. Nenhuma gravação na camada Prata.")

# df_dim_comarca.show(5)







# CARREGAR PARA CAMADA OURO

tb_dim_comarca_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Base/tb_dim_comarca'
# tb_dim_comarca_slv = 'LH_SLV_Produtividade.Base.tb_dim_comarca'
tb_dim_comarca_gld_path = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_GLD_Produtividade.Lakehouse/Tables/Base/tb_dim_comarca'
tb_dim_comarca_gld = 'LH_GLD_Produtividade.Base.tb_dim_comarca' 


if spark.catalog._jcatalog.tableExists(tb_dim_comarca_gld):
    spark.sql(f"DROP TABLE {tb_dim_comarca_gld}")
    print(f"✅ Tabela '{tb_dim_comarca_gld}' foi excluída.")
else:
    print(f"ℹ️ Tabela '{tb_dim_comarca_gld}' não existe.")

dt = DeltaTable.forPath(spark, tb_dim_comarca_slv)
df_dados = dt.toDF()
table_path = tb_dim_comarca_gld_path
df_dados.write.format("delta").mode("overwrite").save(table_path)

```

```python
##TB_FATO_PRODUTIVIDADE

from pyspark.sql import SparkSession
from pyspark.sql.functions import col, when, rank, lit, lpad
from pyspark.sql.window import Window
from delta.tables import DeltaTable
from pyspark.sql.utils import AnalysisException
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, LongType
from pyspark.sql import functions as F
import shutil
import os
from notebookutils import mssparkutils


# Usar a variável nome_workspace que já foi definida na célula 3
if 'nome_workspace' not in dir():
    nome_workspace = notebookutils.notebook.getParameterValue("nome_workspace")
    if not nome_workspace:
        raise ValueError("❌ ERRO: Variável 'nome_workspace' não está definida!")

print(f"✅ TB_FATO_PRODUTIVIDADE - Usando workspace: {nome_workspace}")

# Caminhos
path_vinculo = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Produtividade/siarhes_vinculo.csv'
path_provimento = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Produtividade/siarhes_provimento.csv'
path_designacao = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Produtividade/siarhes_designacao.csv'
path_setor_funcionario = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Produtividade/siarhes_setor_funcionario.csv'
tb_fato_produtividade_bnz = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Tables/Produtividade/tb_fato_produtividade'

# path_tarefas = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Produtividade/tarefas_01012025_-_31052025.gz'
path_tarefas = None
# directory_to_scan = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Produtividade/Lidos/Tarefas/'
directory_to_scan = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Files/Produtividade/Tarefas/'



#Transformar tarefas e carga das tarefas na tb_fato_produtividade em bronze
def ler_csv_com_renomeacao(path, renomear=None):
    df = spark.read.option("header", True).csv(path)
    if renomear:
        for antigo, novo in renomear.items():
            df = df.withColumnRenamed(antigo, novo)
    return df

def aplicar_rank_e_filtrar(df, part_col, order_col, ordem='desc'):
    w = Window.partitionBy(part_col).orderBy(col(order_col).desc() if ordem == 'desc' else col(order_col).asc())
    return df.withColumn("versao_ordem_inversa", rank().over(w)).filter(col("versao_ordem_inversa") == 1)

def preparar_df_tarefa(path):
    # A leitura agora usa o schema definido e ignora o cabeçalho.
    return spark.read.option("header", False).schema(task_schema).csv(path)



def filtrar_por_cpfs(df_base, df_filtrar):
    cpfs = [row["cpf_usuario"] for row in df_filtrar.select("cpf_usuario").distinct().collect()]
    return df_base.filter(col("cpf_usuario").isin(cpfs))

#Função para padronizar CPFs com 11 dígitos
def padronizar_cpfs(df, coluna_cpf="cpf"):
    # Preenche com zeros à esquerda onde tiver menos de 11 dígitos
    df = df.withColumn(coluna_cpf, lpad(col(coluna_cpf), 11, "0"))
    return df

def adicionar_coluna_vinculo(df):
    return df.withColumn("in_existe_vinculo", col("dt_fim").between(col("dtexerc"), col("dtvac")))

# def adicionar_campo_id_produtividade(df):
#     w = Window.orderBy("cd_sistema_origem")
#     return df.withColumn("id_produtividade", rank().over(w).cast("string"))

def main():
    global spark
    spark = SparkSession.builder.getOrCreate()




print(f"Verificando o conteúdo do diretório: {directory_to_scan}")

try:
    # Apenas verifica se há algum arquivo no diretório
    directory_contents = mssparkutils.fs.ls(directory_to_scan)
    
    files_found = [item for item in directory_contents if item.isFile and (item.name.lower().endswith('.csv') or item.name.lower().endswith('.gz'))]

    if files_found:
        path_tarefas = f"{directory_to_scan}*.{{csv,gz}}"
        
        print(f"Sucesso: {len(files_found)} arquivo(s) (.csv ou .gz) encontrado(s).")
        print(f"Caminho de leitura definido para processar todos os arquivos: {path_tarefas}")
    else:
        path_tarefas = None
        print("AVISO: Nenhum arquivo .csv ou .gz encontrado no diretório. O processo será interrompido.")

except Exception as e:
    path_tarefas = None
    print(f"ERRO: Não foi possível acessar o diretório '{directory_to_scan}'. Detalhes: {e}")

# A lógica a seguir para usar path_tarefas permanece a mesma
if path_tarefas:
    print("\nIniciando a leitura dos arquivos...")
else:
    print("\nNenhum arquivo para processar. Finalizando a execução.")



task_schema = StructType([
    StructField("id_", StringType(), True),
    StructField("dt_inicio", StringType(), True),
    StructField("dt_fim", StringType(), True),
    StructField("task_", StringType(), True),
    StructField("id_orgao_julgador", StringType(), True),
    StructField("nr_processo", StringType(), True),
    StructField("cpf_usuario", StringType(), True),
    StructField("tipo", StringType(), True)
])


df_tarefa = preparar_df_tarefa(path_tarefas)

df_vinculo = ler_csv_com_renomeacao(path_vinculo, {"cpf": "cpf_usuario"})
df_vinculo = df_vinculo.withColumnRenamed("cpf", "cpf_usuario")
df_vinculo = padronizar_cpfs(df_vinculo, "cpf_usuario")
df_vinculo = filtrar_por_cpfs(df_vinculo, df_tarefa)
df_tarefa = df_tarefa.join(df_vinculo.select("cpf_usuario", "tipovinc", "dtexerc", "dtvac"), on="cpf_usuario")
df_tarefa = adicionar_coluna_vinculo(df_tarefa)
df_tarefa = aplicar_rank_e_filtrar(df_tarefa, "id_", "dtexerc")

df_provimento = ler_csv_com_renomeacao(path_provimento, {"cpf": "cpf_usuario"})
df_provimento = padronizar_cpfs(df_provimento, "cpf_usuario")
df_provimento = filtrar_por_cpfs(df_provimento, df_tarefa)
df_tarefa = df_tarefa.join(
        df_provimento.select("cpf_usuario", "cargo", col("dtini").alias("dtini_provimento"), col("dtfim").alias("dtfim_provimento")),
        on="cpf_usuario", how="left"
    ).fillna({'cargo': '0', 'dtini_provimento': '1900-00-00 00:00:00', 'dtfim_provimento': '2500-00-00 00:00:00'})
df_tarefa = aplicar_rank_e_filtrar(df_tarefa, "id_", "dtini_provimento")

df_designacao = ler_csv_com_renomeacao(path_designacao, {"cpf": "cpf_usuario"})
df_designacao = padronizar_cpfs(df_designacao, "cpf_usuario")
df_designacao = filtrar_por_cpfs(df_designacao, df_tarefa)
df_tarefa = df_tarefa.join(
        df_designacao.select("cpf_usuario", "funcao", col("dtini").alias("dtini_funcao"), col("dtfim").alias("dtfim_funcao")),
        on="cpf_usuario", how="left"
    ).fillna({'funcao': '0', 'dtini_funcao': '1900-00-00 00:00:00', 'dtfim_funcao': '2500-00-00 00:00:00'})
df_tarefa = aplicar_rank_e_filtrar(df_tarefa, "id_", "dtini_funcao")

df_setor_funcionario = ler_csv_com_renomeacao(path_setor_funcionario, {"cpf": "cpf_usuario"})
df_setor_funcionario = padronizar_cpfs(df_setor_funcionario, "cpf_usuario")
df_setor_funcionario = filtrar_por_cpfs(df_setor_funcionario, df_tarefa)
df_tarefa = df_tarefa.join(
        df_setor_funcionario.select("cpf_usuario", "setor", col("dtini").alias("dtini_setor"), "origem", "prioridade"),
        on="cpf_usuario", how="left"
    ).fillna({'setor': '0', 'dtini_setor': '1900-00-00 00:00:00', 'prioridade': '1'})

df_tarefa = aplicar_rank_e_filtrar(df_tarefa, "id_", "dtini_setor")
df_tarefa = aplicar_rank_e_filtrar(df_tarefa, "id_", "prioridade", ordem='asc')

df_tarefa = df_tarefa.select(
        col("id_").alias("cd_sistema_origem"),
        col("dt_inicio"),
        col("dt_fim"),
        col("task_").alias("ds_tarefa"),
        col("id_orgao_julgador").alias("cd_unidade_tarefa_pje_1g"),
        col("nr_processo"),
        col("tipo"),
        col("cpf_usuario").alias("cpf"),
        col("tipovinc").alias("ds_vinculo"),
        col("cargo").alias("cd_sistema_origem_cargo"),
        col("funcao").alias("cd_sistema_origem_funcao"),
        col("setor").alias("cd_sistema_origem_unidade")
    )

df_tarefa = df_tarefa.withColumn("cd_sistema_origem_funcao", col("cd_sistema_origem_funcao").cast("long"))
df_tarefa = df_tarefa.withColumn("cd_sistema_origem_cargo", col("cd_sistema_origem_cargo").cast("long"))
df_tarefa = df_tarefa.withColumn("cd_sistema_origem_unidade", col("cd_sistema_origem_unidade").cast("long"))

# display(df_tarefa)
# df_tarefa = adicionar_campo_id_produtividade(df_tarefa)
# df_tarefa.write.format("delta").mode("overwrite").option("overwriteSchema", "true").save(tb_fato_produtividade_bnz)
# Verifica se a tabela Delta já existe
if DeltaTable.isDeltaTable(spark, tb_fato_produtividade_bnz):
    # Carrega a tabela Delta existente
    # delta_table = DeltaTable.forPath(spark, tb_fato_produtividade_bnz)

    # (
    #     delta_table.alias("target")
    #     .merge(
    #         df_tarefa.alias("source"),
    #         "target.id_produtividade = source.id_produtividade"
    #     )
    #     .whenMatchedUpdateAll()
    #     .whenNotMatchedInsertAll()
    #     .execute()
    # )
    df_tarefa.write.format("delta").mode("append").option("mergeSchema", "true").save(tb_fato_produtividade_bnz)
    # print("Merge TB_Produtividade")
else:
    # Caso não exista, escreve a tabela Delta
    df_tarefa.write.format("delta").mode("overwrite").save(tb_fato_produtividade_bnz)
    # print("Criação TB_Produtividade")


# CARREGAR PARA CAMADA SILVER

tb_fato_produtividade_bnz = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_BNZ_Produtividade.Lakehouse/Tables/Produtividade/tb_fato_produtividade'

tb_dim_tarefas_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Produtividade/tb_dim_tarefas'
tb_dim_pessoa_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Base/tb_dim_pessoa'
tb_dim_unidade_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Base/tb_dim_unidade'
tb_dim_funcao_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Base/tb_dim_funcao'
tb_dim_cargo_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Base/tb_dim_cargo'
tb_dim_vinculo_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Base/tb_dim_vinculo'
tb_dim_tempo_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Base/tb_dim_tempo'
tb_fato_produtividade_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Produtividade/tb_fato_produtividade'

#%% Carregar dados usando spark.read.format("delta")
df_dados = spark.read.format("delta").load(tb_fato_produtividade_bnz)
df_tarefa = spark.read.format("delta").load(tb_dim_tarefas_slv).select('id_tarefa', 'ds_tarefa', 'peso_tarefa')
df_pessoa = spark.read.format("delta").load(tb_dim_pessoa_slv).select('id_pessoa', 'ds_cpf')
df_pessoa = df_pessoa.withColumnRenamed('ds_cpf', 'cpf')

df_unidade = spark.read.format("delta").load(tb_dim_unidade_slv)
df_unidade = df_unidade.select('id_unidade', 'cd_sistema_origem', 'cd_pje_1g')
df_unidade = df_unidade.withColumn('id_unidade_pessoa', F.col('id_unidade').cast('double'))
df_unidade = df_unidade.select(
    F.col('id_unidade').alias('id_unidade_processo'),
    F.col('id_unidade_pessoa'),
    F.col('cd_sistema_origem').alias('cd_sistema_origem_unidade'),
    F.col('cd_pje_1g').alias('cd_unidade_tarefa_pje_1g')
)



df_funcao = spark.read.format("delta").load(tb_dim_funcao_slv).select('id_funcao', 'cd_sistema_origem')
df_funcao = df_funcao.withColumnRenamed('cd_sistema_origem', 'cd_sistema_origem_funcao')

df_cargo = spark.read.format("delta").load(tb_dim_cargo_slv).select('id_cargo', 'cd_sistema_origem')
df_cargo = df_cargo.withColumnRenamed('cd_sistema_origem', 'cd_sistema_origem_cargo')

df_vinculo = spark.read.format("delta").load(tb_dim_vinculo_slv)
df_tempo = spark.read.format("delta").load(tb_dim_tempo_slv)

#%% Ajustes de tipo e limpeza
df_dados = df_dados.withColumn("cpf", F.col("cpf").cast("string"))
df_pessoa = df_pessoa.withColumn("cpf", F.col("cpf").cast("string"))

df_dados = df_dados.withColumn("cd_sistema_origem_unidade", F.col("cd_sistema_origem_unidade").cast("string"))
df_unidade = df_unidade.withColumn("cd_sistema_origem_unidade", F.col("cd_sistema_origem_unidade").cast("string"))

df_dados = df_dados.withColumn("nr_anomes", F.date_format(F.col("dt_fim"), "yyyyMM").cast("int"))
df_tempo = df_tempo.withColumn("nr_anomes", F.col("nr_anomes").cast("int"))

df_tarefa = df_tarefa.withColumn("peso_tarefa", F.col("peso_tarefa").cast("float"))
df_dados = df_dados.withColumn("nr_processo", F.regexp_replace("nr_processo", "\\D", ""))


# Garantir que ambas colunas são strings para join
df_dados = df_dados.withColumn("cd_unidade_tarefa_pje_1g", F.col("cd_unidade_tarefa_pje_1g").cast("string"))
df_unidade = df_unidade.withColumn("cd_unidade_tarefa_pje_1g", F.col("cd_unidade_tarefa_pje_1g").cast("string"))



#%% Realizar joins (equivalente merges) 
df_dados = df_dados.join(df_tarefa, on=['ds_tarefa'], how='inner')
df_dados = df_dados.join(df_pessoa, on=['cpf'], how='inner')
df_dados = df_dados.join(df_unidade.select('id_unidade_processo', 'cd_unidade_tarefa_pje_1g'), on=['cd_unidade_tarefa_pje_1g'], how='left')



df_dados = df_dados.withColumn(
    'id_unidade_processo',
    F.when(F.col('id_unidade_processo').isNull(), F.lit(1)).otherwise(F.col('id_unidade_processo'))
)

df_dados = df_dados.join(df_unidade.select('id_unidade_pessoa', 'cd_sistema_origem_unidade'), on=['cd_sistema_origem_unidade'], how='left')
df_dados = df_dados.withColumn(
    'id_unidade_pessoa',
    F.when(F.col('id_unidade_pessoa').isNull(), F.lit(1)).otherwise(F.col('id_unidade_pessoa'))
)

df_dados = df_dados.join(df_funcao, on=['cd_sistema_origem_funcao'], how='left')
df_dados = df_dados.withColumn(
    'id_funcao',
    F.when(F.col('id_funcao').isNull(), F.lit(1)).otherwise(F.col('id_funcao'))
)

df_dados = df_dados.join(df_cargo, on=['cd_sistema_origem_cargo'], how='left')
df_dados = df_dados.withColumn(
    'id_cargo',
    F.when(F.col('id_cargo').isNull(), F.lit(1)).otherwise(F.col('id_cargo'))
)

df_dados = df_dados.join(df_vinculo, on=['ds_vinculo'], how='inner')
df_dados = df_dados.join(df_tempo, on=['nr_anomes'], how='inner')


#%% Selecionar colunas desejadas
df_dados = df_dados.select(
    'nr_processo', 'nr_anomes', 'id_tarefa', 'peso_tarefa', 'id_pessoa', 'id_unidade_processo',
    'id_unidade_pessoa', 'id_funcao', 'id_cargo', 'id_vinculo', 'id_tempo', 'tipo'
)

#%% Agrupar e agregar
df_dados = df_dados.groupBy(
    'nr_processo', 'nr_anomes', 'id_tarefa', 'peso_tarefa', 'id_pessoa', 'id_unidade_processo',
    'id_unidade_pessoa', 'id_funcao', 'id_cargo', 'id_vinculo', 'id_tempo', 'tipo'
).agg(
    F.count('nr_processo').alias('qt_tarefa')
)

#%% Criar coluna id_produtividade com expressão complexa
from pyspark.sql.functions import concat_ws, col

df_dados = df_dados.withColumn(
    'id_produtividade',
    F.expr(
        "concat_ws('_', nr_processo, id_pessoa, id_unidade_pessoa, id_unidade_processo, id_funcao, id_cargo, id_vinculo, id_tempo, id_tarefa, qt_tarefa)"
    )
)

#%% Análise de duplicadas
df_ids_duplicados = df_dados.groupBy("id_produtividade").count().filter(col("count") > 1)

if df_ids_duplicados.count() > 0:
    print("--- Linhas duplicadas encontradas com base em 'id_produtividade' ---")
    df_linhas_duplicadas = df_dados.join(
        F.broadcast(df_ids_duplicados),
        on="id_produtividade",
        how="inner"
    ).sort("id_produtividade")
    df_linhas_duplicadas.show(truncate=False, vertical=True)
    print("--- Fim da análise de duplicadas ---")
else:
    print("Nenhuma duplicata encontrada. O MERGE pode prosseguir.")

#%% Deduplicar e realizar MERGE
df_dados_deduplicado = df_dados.dropDuplicates(['id_produtividade'])

# Verifica se a Delta Table já existe
if DeltaTable.isDeltaTable(spark, tb_fato_produtividade_slv):
    delta_table = DeltaTable.forPath(spark, tb_fato_produtividade_slv)
    delta_table.alias("target").merge(
        source=df_dados_deduplicado.alias("source"),
        condition="target.id_produtividade = source.id_produtividade"
    ).whenMatchedUpdateAll().whenNotMatchedInsertAll().execute()
    print("Merge realizado na TB_FATO_PRODUTIVIDADE")
else:
    df_dados_deduplicado.write.format("delta").mode("overwrite").save(tb_fato_produtividade_slv)
    print("Write realizado na TB_FATO_PRODUTIVIDADE")







#CARREGAR PARA CAMADA OURO

tb_fato_produtividade_slv = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_SLV_Produtividade.Lakehouse/Tables/Produtividade/tb_fato_produtividade'
# tb_fato_produtividade_bnz = 'LH_SLV_Produtividade.Produtividade.tb_fato_produtividade'
tb_fato_produtividade_gld_path = f'abfss://{nome_workspace}@onelake.dfs.fabric.microsoft.com/LH_GLD_Produtividade.Lakehouse/Tables/Produtividade/tb_fato_produtividade'
tb_fato_produtividade_gld = 'LH_GLD_Produtividade.Produtividade.tb_fato_produtividade' 


#tb_fato_produtividade
if spark.catalog._jcatalog.tableExists(tb_fato_produtividade_gld):
    spark.sql(f"DROP TABLE {tb_fato_produtividade_gld}")
    print(f"✅ Tabela '{tb_fato_produtividade_gld}' foi excluída.")
else:
    print(f"ℹ️ Tabela '{tb_fato_produtividade_gld}' não existe.")

# dt = DeltaTable(tb_fato_produtividade_slv)
dt = DeltaTable.forPath(spark, tb_fato_produtividade_slv)
# df_dados = dt.to_pandas()
df_dados = dt.toDF()
table_path = tb_fato_produtividade_gld_path
# write_deltalake(table_path, df_dados, mode='overwrite')
df_dados = df_dados.withColumn("id_unidade_processo", col("id_unidade_processo").cast("float"))
df_dados = df_dados.withColumn("id_unidade_pessoa", col("id_unidade_pessoa").cast("float"))
df_dados.write.format("delta").mode("overwrite").save(table_path)


```
